```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
 
# Set data, relating to each statistical measure
df = pd.DataFrame({
'group': ['eta','gamma'],
'1': [-0.1845,0.5947],
'2': [-0.6338,0.8285],
'3': [-0.1391,-0.4424],
'4': [0.4925,0.2491],
'5': [ -0.0460,0.4340],
'6': [0.0206,0.4001],
'7': [0.0531,-0.1602],
'8': [-0.5767,0.6689],
'9': [-0.2374,-0.2543],
'10': [-0.1627,0.2081],
'11': [-0.1799,0.2217],
'12': [-0.1979,0.1871],
'13': [-0.0296,-0.0179],
'14': [-0.1685, -0.0146],
'15': [-0.1259, 0.0222],
'16': [0.0012,-0.0861],
'17': [-0.1110,0.0519],
})
```

```python
# Define a function that forms the plot
def make_spider( row, title, color):
    # number of variable
    categories=list(df)[1:]
    N = len(categories)
    
    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    
    # Initialise the spider plot
    ax = plt.subplot(4,2,row+1, polar=True, )
    
    # Edit spacing on subplots
    plt.subplots_adjust(left=1, bottom=3, right=2, top=4, wspace=0.2, hspace=0.4)

    # If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
    
    # Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='black', size=10)
    
    # Draw ylabels
    ax.set_rlabel_position(0)
    #plt.yticks([1,2,3], ["-.5",".5","1.5"], color="grey", size=7)
    plt.yticks([-1,0,1], ["","",""], color="black", size=5)
    plt.ylim(-1,1)
    
    # Ind1
    values=df.loc[row].drop('group').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.4)
    
    # Add a title
    plt.title(title, size=1, color=color, y=1.1)
```

```python
# initialize the figure
my_dpi=96
plt.figure(figsize=(1000/my_dpi, 1000/my_dpi), dpi=my_dpi)
 
# Create a color palette:
my_palette = plt.cm.get_cmap("Set2", len(df.index))

# Loop to plot
for row in range(0, len(df.index)):
    #make_spider( row=row, title='Node '+df['group'][row], color=my_palette(row))
    make_spider( row=row, title='', color=my_palette(row))
```
